### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling for mismatched lengths
if len(x) != len(y):
    raise ValueError(f"Length of x: {len(x)} and y: {len(y)} do not match. Both must have the same length.")

# Error handling for non-numeric data in x and y
if not all(isinstance(i, (int, float)) for i in x):
    raise TypeError("All values in the 'x' list must be numeric (either int or float).")
if not all(isinstance(i, (int, float)) for i in y):
    raise TypeError("All values in the 'y' list must be numeric (either int or float).")

# Create a scatter plot
plt.scatter(x, y)

# Annotate each point with its (x, y) value using plt.annotate()
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', 
                 (x[i], y[i]), 
                 textcoords="offset points", 
                 xytext=(5,5),  # Offset the annotation slightly
                 ha='center', 
                 fontsize=9)

# Labeling the axes and the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Annotations')

# Show the plot
plt.show()
```
