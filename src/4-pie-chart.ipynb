### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling for mismatched lengths
if len(labels) != len(sizes):
    raise ValueError(f"Length of labels: {len(labels)} and sizes: {len(sizes)} do not match. Both must have the same length.")

# Error handling for non-numeric data in sizes
if not all(isinstance(i, (int, float)) for i in sizes):
    raise TypeError("All values in the 'sizes' list must be numeric (either int or float).")

# Error handling for negative values in sizes
if any(i < 0 for i in sizes):
    raise ValueError("All values in the 'sizes' list must be non-negative.")

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Title
plt.title('Pie Chart')

# Show the plot
plt.show()
```
